2. Create the Streamlit Application (e.g., app.py):
Write your Streamlit application code that utilizes Presidio for PII detection and anonymization. This example demonstrates a basic application.

```python
%%writefile app.py
import streamlit as st
from presidio_analyzer import AnalyzerEngine
from presidio_anonymizer import AnonymizerEngine

st.title("Presidio PII Detection & Anonymization Demo")

text_input = st.text_area("Enter text for PII analysis:", height=200)

if st.button("Analyze and Anonymize"):
    if text_input:
        analyzer = AnalyzerEngine()
        anonymizer = AnonymizerEngine()

        # Analyze for PII
        results = analyzer.analyze(text=text_input, language='en')

        # Anonymize the text
        anonymized_text = anonymizer.anonymize(text=text_input, analyzer_results=results)

        st.subheader("Detected PII Entities:")
        if results:
            for res in results:
                st.write(f"- {res.entity_type} at position {res.start}-{res.end} (Score: {res.score:.2f})")
        else:
            st.write("No PII entities detected.")

        st.subheader("Anonymized Text:")
        st.write(anonymized_text.text)
    else:
        st.warning("Please enter some text to analyze.")
```
